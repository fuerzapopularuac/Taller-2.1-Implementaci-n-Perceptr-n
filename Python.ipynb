```python
import math
import random

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=20, activation="step"):
        self.weights = [random.uniform(-1, 1) for _ in range(input_size)]
        self.bias = random.uniform(-1, 1)
        self.lr = learning_rate
        self.epochs = epochs
        self.activation_name = activation

    # 🔹 Funciones de activación
    def activation(self, x):
        if self.activation_name == "linear":
            return x
        elif self.activation_name == "step":
            return 1 if x >= 0 else 0
        elif self.activation_name == "sigmoid":
            return 1 / (1 + math.exp(-x))
        elif self.activation_name == "relu":
            return max(0, x)
        elif self.activation_name == "tanh":
            return math.tanh(x)
        elif self.activation_name == "softmax":
            return math.exp(x) / (math.exp(x) + math.exp(-x))
        else:
            raise ValueError("Función no válida")

    # 🔹 Predicción
    def predict(self, inputs):
        summation = sum(w * x for w, x in zip(self.weights, inputs)) + self.bias
        return self.activation(summation)

    # 🔹 Entrenamiento
    def train(self, X, y):
        for epoch in range(self.epochs):
            print(f"\nEpoch {epoch+1}")
            for inputs, target in zip(X, y):
                output = self.predict(inputs)

                # Redondeo para funciones continuas
                if self.activation_name != "step":
                    output = round(output)

                error = target - output

                # Actualización de pesos
                for i in range(len(self.weights)):
                    self.weights[i] += self.lr * error * inputs[i]

                self.bias += self.lr * error

                print(f"Entrada: {inputs}, Esperado: {target}, Predicho: {output}")

        print("\nPesos finales:", self.weights)
        print("Bias final:", self.bias)


# ===============================
# 🧪 CASOS DE PRUEBA
# ===============================

# 1. AND
X_and = [[0,0],[0,1],[1,0],[1,1]]
y_and = [0,0,0,1]

# 2. OR
X_or = [[0,0],[0,1],[1,0],[1,1]]
y_or = [0,1,1,1]

# 3. SPAM
X_spam = [[0,0],[1,0],[0,1],[1,1]]
y_spam = [0,1,1,1]

# 4. CLIMA
X_clima = [[0,0],[0,1],[1,0],[1,1]]
y_clima = [0,0,1,1]

# 5. FRAUDE
X_fraude = [[0,0],[1,0],[0,1],[1,1]]
y_fraude = [0,1,1,1]

# 6. RIESGO ACADÉMICO (NORMALIZADO)
X_riesgo = [
    [10/20],
    [12/20],
    [13/20],
    [14/20],
    [16/20],
    [18/20]
]

y_riesgo = [1,1,1,0,0,0]


# ===============================
# 🚀 EJECUCIÓN GENERAL
# ===============================

casos = [
    ("AND", X_and, y_and, "step"),
    ("OR", X_or, y_or, "step"),
    ("SPAM", X_spam, y_spam, "sigmoid"),
    ("CLIMA", X_clima, y_clima, "tanh"),
    ("FRAUDE", X_fraude, y_fraude, "relu"),
    ("RIESGO ACADEMICO", X_riesgo, y_riesgo, "step"),
]

for nombre, X, y, act in casos:
    print(f"\n==============================")
    print(f"CASO: {nombre} | Activación: {act}")
    print(f"==============================")

    p = Perceptron(input_size=len(X[0]), learning_rate=0.1, epochs=20, activation=act)
    p.train(X, y)

    print("\n🔍 Pruebas finales:")
    for x in X:
        pred = p.predict(x)
        if act != "step":
            pred = round(pred)
        print(f"{x} -> {pred}")
```

```python
print("\n===== PRUEBA DE TODAS LAS ACTIVACIONES =====")

activaciones = ["linear", "step", "sigmoid", "relu", "tanh", "softmax"]

for act in activaciones:
    print(f"\nActivación: {act}")
    p = Perceptron(2, activation=act, epochs=15)
    p.train([[0,0],[1,1]], [0,1])
```
